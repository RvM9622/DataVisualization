```python
import pandas as pd
import numpy as np

df = pd.read_csv('flight_1.csv')
df
```

```python
formats = ['%m-%d-%Y', '%d-%m-%Y', '%m/%d/%Y', '%d/%m/%Y', '%Y-%m-%d', '%Y/%m/%d', '%Y-%d-%m', '%Y/%d/%m']
```

```python
from datetime import datetime

def parse_date(date):
    for fmt in formats:
        try:
            return datetime.strptime(date, fmt)  # Try each format
        except ValueError:
            continue
    return None
```

```python
df['FFP_DATE'] = df['FFP_DATE'].apply(parse_date)
df
```

```python
df['FIRST_FLIGHT_DATE'] = df['FIRST_FLIGHT_DATE'].apply(parse_date)
df
```

```python
df['FFP_DATE'] = np.where(df['FFP_DATE'].isna(), df['FIRST_FLIGHT_DATE'], df['FFP_DATE'])
df
```

```python
df['FIRST_FLIGHT_DATE'] = np.where(df['FIRST_FLIGHT_DATE'].isna(), df['FFP_DATE'], df['FIRST_FLIGHT_DATE'])
df
```

```python
df['WORK_PROVINCE'] = np.where(df['WORK_PROVINCE'].isna(), df['WORK_CITY'], df['WORK_PROVINCE'])
df
```

# D. Time-Based Analysis

# 16. Membership Sign-ups Over the Years

```python
plt.figure(figsize=(12, 6))
df['FFP_YEAR'] = df['FFP_DATE'].dt.year
sns.countplot(data=df, x='FFP_YEAR', palette='pastel')
plt.title('Membership Sign-ups Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

# 17. First Flight Date Trends

```python
plt.figure(figsize=(12, 6))
df['FIRST_FLIGHT_YEAR'] = df['FIRST_FLIGHT_DATE'].dt.year
sns.lineplot(data=df.groupby('FIRST_FLIGHT_YEAR').size(), marker='o', color='blue')
plt.title('First Flight Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.show()
```

# 18. Cumulative Points Earned Over the Years

```python
plt.figure(figsize=(12, 6))
cumulative_points = df.sort_values('FFP_DATE').groupby('FFP_YEAR')['Points_Sum'].sum().cumsum()
sns.lineplot(x=cumulative_points.index, y=cumulative_points.values, marker='o', color='purple')
plt.title('Cumulative Points Earned Over the Years')
plt.xlabel('Year')
plt.ylabel('Cumulative Points')
plt.grid(True)
plt.show()
```
